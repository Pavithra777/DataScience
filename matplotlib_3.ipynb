```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
fifa = pd.read_csv('fifa_data.csv')
fifa
```

```python
bins=[40,50,60,70,80,90,100]
plt.title('Player skills distribution')
plt.hist(fifa.Overall,bins = bins,color='#abcdef')

plt.show()
```

```python
fifa.columns
```

```python
right= (fifa.loc[fifa['Preferred Foot']=='Right'].count())[0]
left= (fifa.loc[fifa['Preferred Foot']=='Left'].count())[0]
labels = ['Right', 'Left']
colors=['#abcdef','#aabbcc']
plt.pie([right,left],labels=labels,autopct='% .2f %%',colors=colors)

plt.title('Foot Preferences of FIFA players')
plt.show()
```

```python
fc_brcln=fifa.loc[fifa['Club']=='FC Barcelona']['Overall']
mc=fifa.loc[fifa['Club']=='Manchester City']['Overall']
jv=fifa.loc[fifa['Club']=='Juventus']['Overall']
labels = ['FC Barcelona','Manchester City','Juventus']
plt.boxplot([fc_brcln,mc,jv],labels=labels)
plt.show()
```
