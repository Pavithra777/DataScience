```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

plt.plot(x,y,label='2x',color='r',linewidth='2',marker='.',markersize=10,linestyle='--',markeredgecolor='blue')

plt.title('My first graph!',fontdict={'fontname':'sans-serif','fontsize':12})
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

plt.legend()
plt.show()
```

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

#plt.plot(x,y,label='2x',color='r',linewidth='2',marker='.',markersize=10,linestyle='--',markeredgecolor='blue')
plt.plot(x,y,'r^--',label='2x')
plt.title('My first graph!',fontdict={'fontname':'sans-serif','fontsize':12})
plt.xlabel('X axis')
plt.ylabel('Y axis')
x=np.arange(0,4.5,0.5)
plt.plot(x,x**2,'b',label='x^2')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

plt.legend()
plt.show()
```

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]
plt.plot(x,y,'r^--',label='2x')
plt.title('My first graph!',fontdict={'fontname':'sans-serif','fontsize':12})
plt.xlabel('X axis')
plt.ylabel('Y axis')
x=np.arange(0,4.5,0.5)
plt.plot(x[:4],x[:4]**2,'b.-',label='x^2')
plt.plot(x[3:10],x[3:10]**2,'y.-',label='x^2')
plt.legend()
plt.show()
```

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]
#Resize the graph
plt.figure(figsize=(5,3),dpi=200)
plt.plot(x,y,'r^--',label='2x')
plt.title('My first graph!',fontdict={'fontname':'sans-serif','fontsize':12})
plt.xlabel('X axis')
plt.ylabel('Y axis')
x=np.arange(0,4.5,0.5)
plt.plot(x[:4],x[:4]**2,'b.-',label='x^2')
plt.plot(x[3:10],x[3:10]**2,'y.-',label='x^2')
plt.legend()
plt.savefig('plot.png')
plt.show()
```

```python
values = [1,2,3,4]
labels=['A','B','C','D']
plt.figure(figsize=(5,3))
bar =plt.bar(labels,values)
bar[0].set_hatch('/')
bar[1].set_hatch('o')
bar[2].set_hatch('*')
bar[3].set_hatch('+')
plt.show()
```
