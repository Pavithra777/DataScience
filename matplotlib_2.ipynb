```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
gas = pd.read_csv('gas_prices.csv')
gas
```

```python
plt.title('Gas Price over time',fontdict={'fontweight':'bold','fontsize':18})
plt.plot(gas.Year,gas.Japan,'b.-',label='Japan')
plt.plot(gas.Year,gas.USA,'r.-',label='USA')
plt.plot(gas.Year,gas['South Korea'],'y.-',label='South Korea')
plt.legend()
plt.show()
```

```python
plt.title('Gas Price over time')
for country in gas:
  if(country != 'Year'):
    plt.plot(gas.Year,gas[country],marker='.',label=country)

plt.xticks(gas.Year[::3])

plt.legend(bbox_to_anchor=(1.1, 1.05))
plt.show()
```
